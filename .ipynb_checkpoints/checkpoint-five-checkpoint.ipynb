## Prepare Your Dataframe

Import any libraries you need and create a dataframe.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sam
from sklearn.model_selection import train_test_split

honey = pd.read_excel('honey_clean.xlsx')
honey
```

## Find Correlations

Use either pairplot or a heatmap or both to find the two variables with the strongest correlation in your dataset.

```python
sam.pairplot(honey, x_vars = ['Production (1,000 pounds)', 'Yield per colony (pounds)', 'Honey producing colonies (thousand)'], y_vars = ['Average price per pound (dollars)'], height = 4, aspect = 1, kind = 'scatter')
plt.show()
```

```python
#for honey
sam.set(rc = {'figure.figsize':(15,8)})
sam.heatmap(honey.corr(), cmap = 'cubehelix', annot = True)
plt.show()
```

## Create Your Model

Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
#for honey df
x = honey['Production (1,000 pounds)']
y = honey['Value of production (1,000 dollars)']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.5, test_size = 0.5, random_state = 100)
print(x_train)
print(y_train)
```

```python
import statsmodels.api as sm
# Adding a constant to get an intercept
x_train_sm = sm.add_constant(x_train)


# Fitting the resgression line using 'OLS'
lr = sm.OLS(y_train, x_train_sm).fit()

# Printing the parameters
lr.params
```

```python
lr.summary()
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, 1101.308540 + 1.783691*x_train, 'r')
plt.show()
```

## Error Terms

Finally, plot your error terms!

```python
#for honey
# Predicting y_value using training data of X
y_train_pred = lr.predict(x_train_sm)

# Creating residuals from the y_train data and predicted y_data
res = (y_train - y_train_pred)


# Plotting the histogram using the residual values
fig = plt.figure()
sam.histplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```
